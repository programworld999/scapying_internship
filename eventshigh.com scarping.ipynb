```python
import requests
from bs4 import BeautifulSoup as BS
import json

try:
    page = requests.get("https://www.eventshigh.com/city/kolkata")
#     print("URL is ok!")

except:
    print("Somting wrong in url")
    


soup = BS(page.text, 'html.parser')
scraped_cards = soup.find_all(class_='ga-card-track')
# print(res[1])
events_detail = []
i = 0
for i in range(len(scraped_cards)):
    events_detail.append('https://www.eventshigh.com'+scraped_cards[i].find('a').get('href'))

# ############ =============== testin  ============================================================
# print(events_detail[2])
# details = requests.get(events_detail[2])
# details_soup = BS(details.text, 'html.parser')

# ### ======== title ========= 
# title = details_soup.find(class_="eh-heading ").contents[0].text
# print("title:  "+title)

# ### ========= Date & time ========
# few_details = details_soup.find_all(class_="eh-main-details-desc")
# date_time = few_details[0].contents[0]
# print("date and time: "+date_time)

# ### ========= Price =======
# price = few_details[2].text
# print("price: "+price)

# ### ========= number of people visiting  =======
# people_visit = few_details[3].contents[0].contents[0].text
# print("people_visit: "+people_visit)

# ### ========== location ==========
# location = few_details[1].text
# print("location: "+location)

#####  ==============================================================================================

all_data = []

for i in range(10):
    details = requests.get(events_detail[i])
    details_soup = BS(details.text, 'html.parser')
    title = details_soup.find(class_="eh-heading ").contents[0].text
#     print("title:  "+title)
    
    few_details = details_soup.find_all(class_="eh-main-details-desc")
    
    date_time = few_details[0].contents[0]
#     print("date and time: "+date_time)
    
    price = few_details[2].text
#     print("price: "+price)
    
    people_visit = few_details[3].contents[0].contents[0].text
#     print("people_visit: "+people_visit)
    
    location = few_details[1].text
#     print("location: "+location)
    url = events_detail[i]
#     print(url)
    all_data.append({'title': title, 'date_time': date_time, 'price': price, 'location':location, 'people_visit': people_visit, 'url': url})

    

all_data = json.dumps(all_data)
print(all_data)
```
